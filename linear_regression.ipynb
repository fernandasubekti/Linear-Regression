# Linear Regression

## Do Manually

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2024)
```

```python
def linear_regression(x, m, c):
    return m*x + c
```

```python
# generate some data points
xdata = np.linspace(0, 10, 80)
y = linear_regression(xdata, 3, 4)
y_noise = np.random.normal(loc=0, scale=4.0, size=len(xdata))
ydata = y + y_noise
```

```python
plt.scatter(xdata, ydata, s=10)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# imagine we have a line plot from the data points
y_line_example = 3 * xdata + 1

plt.scatter(xdata, ydata, s=10)
plt.plot(xdata, y_line_example, c="r", label="$y = 3x + 1$")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.plot([5.95, 5.95], [19, 29], color='b', linestyle='--', linewidth=1)

plt.text(5.5, 30, "$(x_i, y_i)$", fontsize=12)
plt.text(5.5, 16, "$(xi, \hat{y_i})$", fontsize=12)
plt.text(5.5, 23, "$r_i$", fontsize=12)

plt.show()
```

```python
# find the best value for m and c
x_bar = sum(xdata) / len(xdata)
y_bar = sum(ydata) / len(ydata)

numerator = 0
denominator = 0
for i in range(len(xdata)):
    numerator += (xdata[i] - x_bar) * ydata[i]
    denominator += (xdata[i] - x_bar) ** 2

m = numerator / denominator
c = y_bar - m * x_bar

print("m value:", round(m, 2))
print("c value:", round(c, 2))
```

```python
# generating predict value
y_pred = m * xdata + c

plt.scatter(xdata, ydata, s=10)
plt.plot(xdata, y_pred, c="r", label="$y = {}x + {}$".format(round(m, 2), round(c, 2)))

plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression Manually")
plt.legend()

plt.show()
```

## Do with the library

```python
from scipy.stats import linregress
```

```python
reg = linregress(xdata, ydata)

print("The value:")
print(reg)
```

```python
y_scipy = reg.slope * xdata + reg.intercept

plt.scatter(xdata, ydata, s=10)
plt.plot(xdata, y_scipy, c="r", label="$y = {}x + {}$".format(round(m, 2), round(c, 2)))

plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression Scipy")
plt.legend()

plt.show()
```
